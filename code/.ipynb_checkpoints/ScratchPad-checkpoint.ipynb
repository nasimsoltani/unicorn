```python
# This Cell calculates the stats for different classes

import pandas as pd
import matplotlib.pyplot as plt
import glob
import pickle
import os

with open (os.path.abspath('../file2app_map.pkl'),'rb') as handle:
    file_app_map = pickle.load(handle)

dataset_base = os.path.abspath('../traces/')
file_list = glob.glob(os.path.join(dataset_base , 'urllc_*'))


dataset_length_dict = {'CallofDuty':[],'Twitch':[],'Zoom':[],'Teams':[],'Facebook':[],'Meet':[]}

for csv_file in file_list:
    this_file = pd.read_csv(csv_file, sep=",").dropna(how='all', axis='columns')
    this_app = file_app_map[csv_file.split('/')[-1].rstrip('.csv')]
    this_length = this_file.shape[0]
    dataset_length_dict[this_app].append(this_length)
    
for app in dataset_length_dict:
    print(app)
    avg = sum(dataset_length_dict[app])/len(dataset_length_dict[app])
    maximum = max(dataset_length_dict[app])
    minimum = min(dataset_length_dict[app])
    empty_string = str(avg)+' '+str(maximum-avg) +' '+str(avg-minimum)
    print("%.2f" %avg, "%.2f" %(maximum-avg), "%.2f" %(avg-minimum))
    # print(empty_string)
    print('---------------------')
```
