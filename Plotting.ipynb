```python
## conf matrix for ORAN dataset

import matplotlib.pyplot as plt
import numpy as np
import matplotlib
import pickle
import os


def plot_confusion_matrix(ax, cm, title='Confusion matrix', cmap=plt.cm.Purples, labels=[], ylabel_insert=False, xlabel_insert=False, bar=False):
    #fig = plt.figure(figsize=(7,7))
    #ax = fig.add_subplot(111)
    im = ax.imshow(cm, interpolation='nearest', cmap=cmap)
    maximum = np.amax(cm)
    for x in range(cm.shape[0]):
        for y in range(cm.shape[1]):
            if cm[x][y]>49:
                ax.annotate(str('{0:.{1}f}'.format(cm[x][y], 0)), xy=(y, x),
                        horizontalalignment='center',
                        verticalalignment='center',fontsize=12,color='w')
            else:
                ax.annotate(str('{0:.{1}f}'.format(cm[x][y], 0)), xy=(y, x),
                        horizontalalignment='center',
                        verticalalignment='center',fontsize=12,color='k')
    
    #plt.title(title,fontsize=24)
    bar_label = ['0','20','40','60','80','100']
#     bar_label = []
    cbar = plt.colorbar(im,fraction=0.046, pad=0.04)
    cbar.ax.set_yticklabels(bar_label)
    cbar.ax.tick_params(labelsize=12) 
    #plt.colorbar()
    tick_marks = np.arange(len(labels))
    
    
    plt.tight_layout()
    ax.set_title(title, fontsize=14)
    if ylabel_insert:
        ax.set_yticks(tick_marks, labels, fontsize=12)
        ax.set_ylabel('True label',fontsize=14)
    else:
        ax.set_yticks(tick_marks, [], fontsize=12)
        
    if xlabel_insert:
        ax.set_xticks(tick_marks, labels, rotation=45, ha='center',fontsize=12)
        ax.set_xlabel('Predicted label',fontsize=14)
    else:
        ax.set_xticks(tick_marks, [], fontsize=12)
    

pickle_list = ['pred_dict-ForwConvNet-oodindex4.pkl',
               'pred_dict-ForwConvNet-oodindex5.pkl',
              'pred_dict-ResConvNet-oodindex4.pkl',
              'pred_dict-ResConvNet-oodindex5.pkl']
conf_list = []

base_path = os.path.abspath('results/') 


ID_classes = ['c1','c2','c3','c4','c5']
for i, pickle_path in enumerate(pickle_list):
    conf = np.zeros((len(ID_classes),len(ID_classes)))
    with open (os.path.join(base_path , pickle_path),'rb') as handle:
        pred_dict = pickle.load(handle)
    out_layer_dict = pred_dict['outlayer']
    
    for label in out_layer_dict:
        this_list = out_layer_dict[label]
        for pred in this_list:
            pred_class = np.argmax(pred[0])
            conf[label,pred_class] += 1 
            
    # # normalize conf [true_class,predicted_class]
    norm_conf = np.zeros(conf.shape)
    for i in range(conf.shape[0]):
        for j in range(conf.shape[1]): # for each predicted
            norm_conf[i,j] = (conf[i,j]/np.sum(conf[i,:]))*100
            
    conf_list.append(norm_conf)

# ---------------------- Now plot -------------------------
y = conf_list
color_list = ['royalblue','red','ForestGreen','orange','black','violet', 'cyan','gray','yellow','salmon','pink']
legends = ['Training Loss','Validation Loss']
markers = ['*','o','+','x']
titles = ['ForwConv\nFirst Set of Classes','ForwConv\nSecond Set of Classes','ResConv\nFirst Set of Classes','ResConv\nSecond Set of Classes']
color_theme_list = [plt.cm.Blues, plt.cm.Blues, plt.cm.Reds, plt.cm.Reds]

plt.rcParams['axes.titley'] = 1    # y is in axes-relative coordinates.
matplotlib.rcParams.update({'font.size': 14})

# fig = plt.figure(figsize=(6,6))
fig = plt.figure(figsize=(12,6))

# print(axes.shape)
print(len(y))
print(len(y[0]))

# subplot_list = [221,222,223,224]
subplot_list = [141,142,143,144]

for i in range(len(y)):
    this_y = y[i]
    title_text = titles[i]
    # ax = axes[i]
    ax = fig.add_subplot(subplot_list[i])
    if i == 0:# or i==2:
        ylabel_insert = True
    else:
        ylabel_insert = False
    if i ==0 or i==2:
        ID_classes=['CallofDuty','Facebook','Meet','Zoom','Teams']
    else:
        ID_classes=['CallofDuty','Facebook','Meet','Zoom','Twitch']
        
    xlabel_insert = True
    # if i == 3:
    #     bar = True
    # else:
    #     bar = False
    
    plot_confusion_matrix(ax=ax, cm=y[i], title=title_text, cmap=color_theme_list[i], labels=ID_classes, ylabel_insert=ylabel_insert, xlabel_insert=xlabel_insert, bar=True) 
   

# fig.tight_layout(pad=1.0)
plt.savefig('conf.pdf', bbox_inches='tight')  
    
```

```python
## read the prediction pickle file and study the pen ultimate layer features
## perform tsne

from sklearn.manifold import TSNE
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import pickle
from matplotlib.pyplot import subplots


ood_index1 = '4'
ood_index2 = '5'
s_size = 1
line_width = 4
titlesize = 14

pickle_list = ['pred_dict-ForwConvNet-oodindex4.pkl',
               'pred_dict-ForwConvNet-oodindex5.pkl',
              'pred_dict-ResConvNet-oodindex4.pkl',
              'pred_dict-ResConvNet-oodindex5.pkl']
base_path = os.path.abspath('results/') 

fig, axes = subplots(figsize=(7,4), nrows=2, ncols=2)
axes = [axes[0,0],axes[0,1],axes[1,0],axes[1,1]]
matplotlib.rcParams.update({'font.size': 12})


for i, pickle_path in enumerate(pickle_list):
    ax = axes[i]
    with open (os.path.join(base_path, pickle_path),'rb') as handle:
        pred_dict = pickle.load(handle)
    
    feature_dict = pred_dict['features']
    feature_dims = 128
    
    stride_list = []
    all_classes_array = np.empty(shape=[0, feature_dims])
    for label in [0,1,2,3,4,5]:    
        this_class_array = np.empty(shape=[0, feature_dims])
        vectored = np.squeeze(np.array(feature_dict[label]))
        this_class_array = np.concatenate((this_class_array, vectored), axis=0)
        stride_list.append(this_class_array.shape[0])
        all_classes_array = np.concatenate((all_classes_array, this_class_array), axis=0)
    
    X = all_classes_array
    print(X.shape)
    
    """for tsne """    
    t_sne = TSNE(n_components=2, init='random')
    
    X_embedded= t_sne.fit_transform(X)
    print(X_embedded.shape)
    
    if i == 0:
        ax.set_title('ForwConv')
    elif i==1:
        ax.set_title('ResConv')
    else: 
        ax.set_title('')
    if i==0 or i==1:
        label_list = ['Call of Duty (ID)','Facebook (ID)','Meet (ID)','Zoom (ID)','Teams (OOD)','Twitch (ID)']
        color_list = ['blue','red','ForestGreen','orange','cyan','magenta']
    elif i==2 or i==3:
        label_list = ['Call of Duty (ID)','Facebook (ID)','Meet (ID)','Zoom (ID)','Teams (ID)','Twitch (OOD)']
        color_list = ['blue','red','ForestGreen','orange','cyan','magenta']
    ax.set_xticklabels([])
    ax.set_yticklabels([])
    if i == 0:
        ax.set_ylabel('First Set\nof Classes', fontsize=titlesize, rotation=90)
    elif i == 2:
        ax.set_ylabel('Second Set\nof Classes', fontsize=titlesize, rotation=90)
    
    # color_list = ['blue','red','ForestGreen','orange','magenta','black', 'cyan','gray','yellow','peru','pink','salmon','purple','brown','lime','sienna','olive','teal']
    # color_list = ['blue','ForestGreen','orange','magenta','cyan', 'red','gray','yellow','peru','pink','salmon','purple','brown','lime','sienna','olive','teal']
    print(len(color_list))
    print(stride_list)
    
    x1 = X_embedded[:stride_list[0],0]
    y1 = X_embedded[:stride_list[0],1]
    start = stride_list[0]
    x2 = X_embedded[start:start+stride_list[1],0]
    y2 = X_embedded[start:start+stride_list[1],1]
    start+= stride_list[1]
    x3 = X_embedded[start:start+stride_list[2],0]
    y3 = X_embedded[start:start+stride_list[2],1]
    start+= stride_list[2]
    x4 = X_embedded[start:start+stride_list[3],0]
    y4 = X_embedded[start:start+stride_list[3],1]
    start+= stride_list[3]
    x5 = X_embedded[start:start+stride_list[4],0]
    y5 = X_embedded[start:start+stride_list[4],1]
    start+= stride_list[4]
    x6 = X_embedded[start:start+stride_list[5],0]
    y6 = X_embedded[start:start+stride_list[5],1]
    
    ax.scatter(x1, y1,s=s_size, linewidths=line_width, color=color_list[0], label=label_list[0])
    ax.scatter(x2, y2,s=s_size, linewidths=line_width, color=color_list[1], label=label_list[1])
    ax.scatter(x3, y3,s=s_size, linewidths=line_width, color=color_list[2], label=label_list[2])
    ax.scatter(x4, y4,s=s_size, linewidths=line_width, color=color_list[3], label=label_list[3])
    ax.scatter(x5, y5,s=s_size, linewidths=line_width, color=color_list[4], label=label_list[4])
    ax.scatter(x6, y6,s=s_size, linewidths=line_width, color=color_list[5], label=label_list[5])
    
    ax.grid()
    if i==1 or i==3:
        ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0, markerscale=2)

    print('-------------------------')

plt.subplots_adjust(wspace=-0.5, hspace=0.1)
fig.tight_layout(pad=1.0)
plt.savefig('clusters.pdf', bbox_inches='tight') 
```
